```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Unsolved/Resources/got.csv")
```

```python
attacker_df = df.value_counts("attacker_king").to_frame().rename(columns={0: "Battles As Attacker"})
```

```python
defender_df = df.value_counts("defender_king").to_frame().rename(columns={0: "Battles As Defender"})
```

```python
merged_df = pd.merge(attacker_df, defender_df, left_index=True, right_index=True, how="outer")
```

```python
merged_df.fillna(0).plot.bar(title="Battle of the Kings", ylabel="Number of Battles", xlabel="King")
```
