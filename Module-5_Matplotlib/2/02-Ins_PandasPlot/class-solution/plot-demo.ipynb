```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "../Resources/avg_rain_state.csv"
```

```python
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
plt.bar(df["State"], df["Inches"])
```

```python
df.sort_values("Inches", ascending=False).plot(x="State", y="Inches", kind="bar", figsize=(20, 10), title="Rainfall by State (inches)", ylabel="Rainfall (inches)")
```

```python
df.sort_values("Inches", ascending=False).plot(x="State", y="Inches", kind="bar", figsize=(20, 10))
plt.title("Rainfall by State (inches)")
plt.ylabel("Rainfall (inches)")
plt.xticks(rotation=-45)
plt.savefig("../Images/fig.png")
plt.show()
```

```python
df.sort_values("Inches", ascending=False).plot(x="State", y=["Inches", "Millimetres" ] , kind="bar", figsize=(20, 10))
```

```python
df.sort_values("Inches", ascending=False).plot(x="State", y=["Inches", "Millimetres" ] , kind="bar", stacked=True, figsize=(20, 10))
```
